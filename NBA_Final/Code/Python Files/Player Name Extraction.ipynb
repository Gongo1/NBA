```
import spacy as nlp
from spacy.tokenizer import Tokenizer
from spacy.lang.en import English
### Aspect Extraction
import spacy 
import pandas as pd
nlp = spacy.load("en_core_web_md")
```

```
from collections import Counter
def get_aspects(article):
    doc = nlp(article)
    aspect_list = []
    ListCount = Counter()
    for chunk in doc.noun_chunks:
        for ent in chunk.ents:
            if ent.label_ in ['NORP','PERSON']:
              aspect_list.append(ent.text)
    return(Counter(list(set(aspect_list))))
```

```
df = pd.read_csv('NBA ESPN Articles.csv')
```

```
ent_list = [get_aspects(article) for article in df.Body]
```

```
len(ent_list)
```

```
df['mention_count'] = pd.Series(ent_list)
df.head()
```
